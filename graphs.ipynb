```python
import numpy as np
import matplotlib.pyplot as plt
from simulation_utils import Simulator
```

## Theoretic Calculation

```python
LAMBDA = 10
MU = 12
Q = 6
RO = LAMBDA/MU

p0 = 1/np.sum(np.array([(np.power(RO, i)) for i in range(Q+1)]))
p0
```

```python
N = p0 * np.sum([i * (np.power(RO, i)) for i in range(Q+1)])
LAMBDA_AVG = p0 * LAMBDA * np.sum([np.power(RO, i) for i in range(Q)])
T_AVG = N/LAMBDA_AVG
print(N, LAMBDA_AVG, T_AVG)
```

## Simulation Mean

```python
T = [10,100,500,1000,2000,2500]
results = {t:[] for t in T}
avg_results = []

for t in T:
    # run simulator with t 1 1 10 5 12
    for _ in range(10):
        simulator = Simulator(t, 1, [1], 10, [5], [12])
        simulator.run()
        results[t].append(simulator.get_total_time())
    
for k,v in results.items():
    avg_results.append(np.mean(v))
```

```python
# scatter plot avg_results vs T
plt.xlabel('T')
plt.ylabel('Average Total Time')
plt.title('Average Total Time vs T')
plt.grid(alpha=0.5)
plt.scatter(T, avg_results)
plt.show()
```
